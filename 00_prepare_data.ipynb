```python
from os import makedirs
# create directories
dataset_home = './_data/dogs_cats/'
subdirs = ['train/', 'test/']
for subdir in subdirs:
    # create label subdirectories
    labeldirs = ['dogs/', 'cats/']
    for labldir in labeldirs:
        newdir = dataset_home + subdir + labldir
        makedirs(newdir, exist_ok=True)
        print("CREATED DIRECTORY:", newdir)
```

```python
from os import makedirs
from os import listdir
from shutil import copyfile
from random import seed
from random import random

# seed random number generator
seed(1)
# define ratio of pictures to use for validation
val_ratio = 0.25
# copy training dataset images into subdirectories
src_directory = './_data/dogs_cats/train'
for file in listdir(src_directory):
    src = src_directory + '/' + file
    dst_dir = 'train/'
    if random() < val_ratio:
        dst_dir = 'test/'
    if file.startswith('cat'):
        dst = dataset_home + dst_dir + 'cats/'  + file
        print(src, dst)
        copyfile(src, dst)
    elif file.startswith('dog'):
        dst = dataset_home + dst_dir + 'dogs/'  + file
        print(src, dst)
        copyfile(src, dst)
```

```python
import torchvision
import torch

from torchvision.models import ResNet50_Weights
weights = ResNet50_Weights.IMAGENET1K_V1
preprocess = weights.transforms()

dataset_train = torchvision.datasets.ImageFolder('./_data/dogs_cats/train', transform=preprocess)
dataset_val = torchvision.datasets.ImageFolder('./_data/dogs_cats/test', transform=preprocess)

print('TRAIN', dataset_train)
print('TEST', dataset_val)
```
