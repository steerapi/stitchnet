```python
from skl2onnx.helpers.onnx_helper import select_model_inputs_outputs
from skl2onnx.helpers.onnx_helper import save_onnx_model
from skl2onnx.helpers.onnx_helper import enumerate_model_node_outputs
from skl2onnx.helpers.onnx_helper import load_onnx_model

from stitchnet.stitchonnx.utils import change_input_dim, get_fragments, Net, get_data_score, load_cats_and_dogs_dset

import numpy as np
from tqdm import tqdm
from glob import glob
from stitchnet.finetuning.finetune import finetune
import os
import torch

num_classes = 2
num_epochs = 10
num_iters = 10
feature_extract = True
batch_size = 32
val_batch_size = 64

data_score,t = get_data_score(32, includeTarget=True)
data_score = list(zip(data_score,t))

data_score = load_cats_and_dogs_dset("train")
```

```python
modelnames = sorted(glob('_models/*.onnx'))
os.makedirs('_results/finetune', exist_ok=True)
for i,modelname in tqdm(enumerate(modelnames)):
    model_onnx1 = load_onnx_model(modelname)
    name = os.path.basename(modelname)
    namewithoutext = os.path.splitext(name)[0]
    for run in range(10):
        os.makedirs(f'_results/finetune/{namewithoutext}', exist_ok=True)
        numsamples, iter_val_acc_history, val_acc_history, val_loss_history, train_acc_history, train_loss_history = finetune(model_onnx1, data_score, num_classes, num_epochs, num_iters, batch_size, val_batch_size, feature_extract)
        # model_ft = model_ft.cpu()
        with open(f'_results/finetune/{namewithoutext}/{run:03}.txt', 'w') as f:
            for n,acc in zip(numsamples, iter_val_acc_history):
                f.write(f'{n},{acc}\n')
        # torch.onnx.export(model_ft, torch.ones(1,3,224,224), f'_results/finetune/{namewithoutext}/{run:03}_ft.onnx')        
        
```
